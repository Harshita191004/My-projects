```python
import matplotlib.pyplot as plt

df['Placed_numeric'] = df['Placed'].map({'Yes': 1, 'No': 0})

scatter = plt.scatter(df['CGPA'], df['IQ_Score'], c=df['Placed_numeric'], cmap='viridis')
plt.xlabel('CGPA')
plt.ylabel('IQ_Score')
plt.title('IQ Score vs CGPA by Placement Status')
plt.colorbar(scatter, ticks=[0, 1], label='Placement Status', format=plt.FuncFormatter(lambda x, p: {0: 'No', 1: 'Yes'}[x]))
plt.show()
```

```python
X=df.iloc[:,0:2]
y=df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
```

```python
X_train
```

```python
X_test=sc.transform(X_test)
```

```python
X_test
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression()
```

```python
clf.fit(X_train,y_train)
```

```python
clf.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,clf.predict(X_test))
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression

# Create a numerical version of y_train for training the classifier for plotting
y_train_num_for_plot = y_train.map({'No': 0, 'Yes': 1})

# Create a new classifier instance for plotting purposes
# This is done because the original 'clf' was trained on string labels
clf_for_plot = LogisticRegression()
clf_for_plot.fit(X_train, y_train_num_for_plot)

# Now use this new classifier for plotting
plot_decision_regions(X_train, y_train_num_for_plot.values, clf=clf_for_plot, legend=2)
```
